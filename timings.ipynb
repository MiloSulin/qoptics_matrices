### plotting ###

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
mus = 10**(-5)
ms = 10**(-3)
```

```python
log_2 = np.array(range(2,12))
mat_size = np.pow(2, log_2)

qo_symp = [229*mus, 289*mus, 351*mus, 438*mus, 776*mus, 3.31*ms, 12.9*ms, 73*ms, 409*ms, 2.38]
qo_pas = [0,0,0,0,0, 3.35*ms, 12.4*ms, 89.4*ms, 338*ms, 2.24]

wl_symp = [184*mus, 193*mus, 215*mus, 280*mus, 525*mus, 2.01*ms, 64.9*ms, 116*ms, 336*ms, 1.45]
wl_pas = [0,0,0,0,0, 645*mus, 33.5*ms, 55.9*ms, 148*ms, 619*ms]
```

```python
fig, ax = plt.subplots(1,2, figsize=(12,4), layout='constrained')
ax[0].scatter(mat_size, qo_symp, label="qoptics_matrices.py")
ax[0].scatter(mat_size, wl_symp, c='r', label="thewalrus")
ax[0].set_xscale('log', base=2)
ax[0].set_yscale('log')
ax[1].scatter(mat_size[5:], qo_pas[5:], label="qoptics_matrices.py")
ax[1].scatter(mat_size[5:], wl_pas[5:], c='r', label="thewalrus")
ax[1].set_xscale('log', base=2)
ax[1].set_yscale('log')
ax[0].legend()
ax[1].legend()
ax[0].set_xlabel("Matriisin koko")
ax[0].set_ylabel("Suoritusajan keskiarvo [s]")
ax[1].set_xlabel("Matriisin koko")
ax[1].set_ylabel("Suoritusajan keskiarvo [s]")
ax[0].set_title("a) Symplektiset matriisit")
ax[1].set_title("b) Ortogonaaliset matriisit")
plt.show()
```
